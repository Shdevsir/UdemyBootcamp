```python
x.add_child(fo.Marker(location=[40, -80], popup='Client', icon=fo.Icon(color='red')))
```

```python
map.add_child(x)
```

```python
volcano = pd.read_csv('239.volcano.csv')
```

```python
lat_vo = list(volcano['Latitude'])
lon_vo = list(volcano['Longitude'])
name_vo = list(volcano['Name'])
```

```python
vol = fo.FeatureGroup(name='New_map')
```

```python
for lat, lon, name in zip(lat_vo, lon_vo, name_vo):
    vol.add_child(fo.Marker(location=[lat, lon], popup=name, icon=fo.Icon(color='blue')))
```

```python
map.add_child(vol)
```
