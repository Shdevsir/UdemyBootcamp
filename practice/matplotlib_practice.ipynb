```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, x, z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('test-graph')
```

```python
COLOR = 'white'
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR
plt.subplot(3, 2, 1)
plt.plot(x, y)
plt.subplot(3, 2, 2)
plt.plot(x, x)
plt.subplot(3, 2, 3)
plt.plot(x, z)
plt.subplot(3, 2, 4)
plt.subplot(3, 2, 5)
plt.subplot(3, 2, 6)
plt.tight_layout()
```

```python
plt.scatter(x, y, label='Example')
plt.legend()
```
